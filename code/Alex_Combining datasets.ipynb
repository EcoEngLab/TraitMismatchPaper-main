This Jupyter combines Cera's dataset with Vectraits into "Fulldataset.csv", then filters out all the traits needed in this project, called "TraitofInterest.csv"

```r
Vec<- read.csv("../data/VecTraits.csv")
subdata <- subset(Vec,Vec$published_data.interactor1phylum=="Arthropoda") #subsetted arthropods here
subdata<-gdata::drop.levels(subdata)

cera<- read.csv("../data/ceramai_digitised_dataset.csv")
```

```r
#standardizing column names
names(subdata) <- gsub("published_data.","",names(subdata)) #removed published_data. from vectraits data
names(subdata) <- gsub("locationtext","location",names(subdata))
cera<- cera[, ! names(cera) %in% c("published","embargorelease"), drop = F]

#changing timestart in Cera's dataset to the secondstressor column to match Vectraits
cera$secondstressorvalue <- cera$timestart
cera$secondstressorunit <- cera$totalobstimeunit
```

```r
#checking if the names match: seem to all match up!
setdiff(names(cera), names(subdata))
```

```r
#changing some of Cera's data so that it has standardized secondary stressor of Time when there is time data
library("dplyr")
cera <- cera %>% mutate(secondstressor = case_when(secondstressorunit %in% c("day","days") ~ 'Time'))
```

```r
data<- rbind(subdata,cera)
str(data)
```

```r
#new data has 15874 rows:
nrow(cera) + nrow(subdata)
```

```r
write.csv(data, file="../data/Fulldataset.csv")
```

```r
#looking at what traits there are:
levels(data$originaltraitname)
```

```r
#subsetting only traits that matter for project:
subset <- subset(data,data$originaltraitname %in% c("Adult longevity (female, bloodfed)", "Adult longevity (male)","Adult survival","Adult survival (female, bloodfed)","Adult survival (male)", "Development Rate", "Development Time", "Development time", "Fecundity","Fecundity Rate","Generation Time", "Juvenile survival","Juvenile survival ","Longevity","Mortality Rate","Oviposition Rate", "Percentage Survival", "Survival Rate","Survivorship", "Survival Time","Egg development time"))
subset<- gdata::drop.levels(subset)

levels(subset$originaltraitname)
levels(subset$interactor1)
```

```r
write.csv(subset,file="../data/TraitofInterest.csv")
```
